```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import scipy.stats as stats
import math
```

```python
#Some inconsistencies between importing scipy as ss vs. stats
import scipy.stats as ss
from scipy.special import factorial
from scipy.stats import beta
from scipy.stats import binom
from scipy.stats import burr
from scipy.stats import chi2
from scipy.stats import gumbel_l
from scipy.stats import f
from scipy.stats import gamma
from scipy.stats import genextreme
from scipy.stats import genpareto
from scipy.stats import geom
from scipy.stats import halfnorm
from scipy.stats import hypergeom
from scipy.stats import lognorm
from scipy.stats import nbinom
from scipy import special
from scipy.stats import t
from scipy.stats import ncx2
from scipy.stats import rayleigh
from scipy.stats import levy_stable
from scipy.stats import randint
from scipy.stats import weibull_min
```

```python
# 1. Normal
mu = 0
variance = 1
sigma = math.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = stats.norm.pdf(x, mu, sigma)

plt.plot(x, y, label = "Normal Distribution", linewidth = 4)
plt.legend()
plt.grid(alpha = 0.4, linestyle = '--')
plt.show

ax = plt.subplot(111)
ax.set_xlabel("Standard deviations")
ax.set_ylabel("Probability Density Function")
ax.set_title("Normal Distribution")
```

```python
# 2. Uniform distribution

y = 1
x = np.arange(0, 1)

a = 0
b = 1

plt.hlines(y, xmin = 0, xmax = 1, linewidth = 4)
plt.hlines(0, xmin = -.5, xmax = a, linewidth = 4)
plt.hlines(0, xmin = b, xmax = 1.5, linewidth = 4)
plt.vlines(a, ymin = 0, ymax = y, linewidth = 4)
plt.vlines(b, ymin = 0, ymax = y, linewidth = 4)

plt.show
plt.grid(alpha=.4,linestyle='--')

#txt_a = "a = " + str(a)
#txt_b = "\nb = " + str(b)
#txt = txt_a + txt_b
#plt.text(x = -0.5, y = 0.8, s = txt, fontsize = 12)

ax = plt.subplot(111)
ax.set_xlabel("Values of x")
ax.set_ylabel("Probability Density Function")
ax.set_title("Uniform Distribution")
```

```python
#4. Poisson Distribution

mu = 5
x = np.arange(0, 20, 0.1)
y = np.exp(-mu)*np.power(mu, x)/factorial(x)

plt.plot(x, y, label = "Poisson Distribution", linewidth = 5)
plt.legend()
plt.grid(alpha = 0.4, linestyle='--')

#txt = r"$\mu$ = " + str(sigma)
#plt.text(x = 15, y = 0.15, s = txt, fontsize = 12)

ax = plt.subplot(111)
ax.set_xlabel("Number of Occurences")
ax.set_ylabel("Probability Density Function")
ax.set_title("Poisson Distribution")

plt.show()

#Not sure if this is better or not
l = ax.fill_between(x, y, color = 'c')
```

```python
# 4. Beta
a = 2
b = 2
x = np.arange (-50, 50, 0.1)
y = beta.pdf(x, a, b, scale=100, loc=-50)

plt.plot(x, y, label = "Beta Distribution", linewidth = 4)
plt.legend()
plt.grid(alpha = 0.4)
plt.show

#txt_a = r"$\alpha$ = " + str(a)
#txt_b = "\n" + r"$\beta$ = " + str(b)
#txt = txt_a + txt_b
#plt.text(x = -10, y = 0.001, s = txt, fontsize = 12)

ax = plt.subplot(111)
ax.set_xlabel("Values of x")
ax.set_ylabel("Probability Density Function")
ax.set_title("Beta Distribution")
```

```python
# 5. Binomial

fig, ax = plt.subplots(1, 1)

n = 20
p = 0.4

mean, var, skew, kurt = binom.stats(n, p, moments='mvsk')

x = np.arange(binom.ppf(0.001, n, p),
              binom.ppf(0.999, n, p))
y = binom.pmf(x, n, p)

plt.plot(x, y, label = "Binomial Distribution", linewidth = 4)
plt.legend()
plt.grid(alpha = 0.4, linestyle = '--')

#txt_p = "p = " + str(p)
#txt_n = "\nn = " + str(n)
#txt = txt_p + txt_n
#plt.text(x = 7, y = 0.025, s = txt, fontsize = 12)

ax = plt.subplot(111)
ax.set_xlabel("Values of x")
ax.set_ylabel("Probability Density Function")
ax.set_title("Binomial Distribution")

plt.show()
```

```python
# 6. Burr

fig, ax = plt.subplots(1, 1)
c, d = 10.5, 4.3
mean, var, skew, kurt = burr.stats(c, d, moments='mvsk')

x = np.linspace(burr.ppf(0.01, c, d),
                burr.ppf(0.99, c, d), 100)
y = burr.pdf(x, c, d)

plt.plot(x, y, label = "Burr Distribution", linewidth = 4)
plt.legend()
plt.grid(alpha = 0.4, linestyle = '--')

ax = plt.subplot(111)
ax.set_xlabel("Values of x")
ax.set_ylabel("Probability Density Function")
ax.set_title("Burr Distribution")

plt.show()
```

```python
# 7. Chi-squared

x = np.arange(0, 20, 0.001)
y = chi2.pdf(x, df = 4)

plt.plot(x, y, label = "Chi-squared distribution", linewidth = 4)
plt.legend()
plt.grid(alpha = 0.4, linestyle = '--')

ax = plt.subplot(111)
ax.set_xlabel("Values of x")
ax.set_ylabel("Probability Density Function")
ax.set_title("Chi-squared Distribution")

plt.show()
```

```python
# 8. Exponential

def plot_exponential(x_range, mu=0, sigma=1, cdf=False, **kwargs):
    x = x_range
    if cdf:
        y = ss.expon.cdf(x, mu, sigma)
    else:
        y = ss.expon.pdf(x, mu, sigma)
    plt.plot(x, y, **kwargs)

mu = 0
sigma = 1
    
x = np.arange(0, 20, 0.001)
y = ss.expon.pdf(x, mu, sigma)

plt.plot(x, y, label = "Exponential Distribution", linewidth = 4)
plt.legend()
plt.grid(alpha = 0.4, linestyle = '--')

ax = plt.subplot(111)
ax.set_xlabel("Values of x")
ax.set_ylabel("Probability Density Function")
ax.set_title("Exponential Distribution")

plt.show()
```
